```python
 # Column from series (in this it don,t give length of vaule doesnot match
#length of index but in python list to column it doesnot give that error)

seats_column = pd.Series([5,5,5,5,5])

#new column called seats
car_sales["Seats"] = seats_column
```

```python
#Column from Python list (in this it can give length of vaule doesnot
#match length of index but in series to column it doesnot give that error)
fuel_economy = [7.5,9.2,5.0,9.6,8.7,4.7,7.6,8.7,3.0,4.5]
car_sales["Fuel  per 100KM"] = fuel_economy
car_sales
```

```python
# column from another column of DataFrame (you can perform operations on column)
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100*car_sales["Fuel  per 100KM"]
car_sales
```

```python
#create a column from single value
car_sales["Number of wheels"] = 4
car_sales
```
